```python
from hh_file import load_yaml , Hodgkin_Huxley , plot_graphs
import numpy as np
```

```python
parameters = load_yaml('parameters.yaml')
```

```python
C_m = parameters['C_m']
E_K = parameters['E_K']
E_L = parameters['E_L']
E_Na = parameters['E_Na']
I_total = parameters['I_total']
V_r = parameters['V_r']
gK_max = parameters['gK_max']
gL = parameters['gL']
gNa_max = parameters['gNa_max']
steps = parameters['steps']
t_span = (0.0,100.0)
time = np.linspace(*t_span, steps)
```

```python
model = Hodgkin_Huxley(t_span=t_span , time=time, steps=steps, V_r=V_r, C_m=C_m, gK_max=gK_max, gNa_max=gNa_max, gL=gL,
                       I_total=I_total, E_K=E_K, E_Na=E_Na, E_L=E_L)

sol = model.solver()
```

```python
V_val=sol[:, 0]
n_val=sol[:, 1]
m_val=sol[:, 2]
h_val=sol[:, 3]
```

```python
plot_graphs(time=time, x1=V_val, x2=n_val, x3=m_val, x4=h_val)
```
